## 各店全品項的總銷售額
+  前25%為A店
+  50%~75%為B店
+  25%~50%為C店
+  後25%為D店,甚至可以淘汰

```python
# 先把'店家代號'column 抓出
Store = df.loc[:,"店家代號"]

# 才能把其他的object型態改成int
dfTotalByStore = df.loc[:,"2022/01":].apply(pd.to_numeric)

# 再concat回來
dfTotalByStore = pd.concat([Store,dfTotalByStore], axis="columns")

dfTotalByStore.head()
```

```python
# 2021~2022/09 全部銷售數量
dfTotalByStore.loc[:,"2022/01":].sum().sum()
```

```python
dfbyStore = dfTotalByStore.groupby("店家代號").sum().sum(axis=1)
dfbyStore.describe()
```

#### 總銷售額為0的店家，去除，極端值，對品牌全年營收沒有任何貢獻

```python
dfbyStore[dfbyStore == 0]
```

#### 店家代號950有些問題，為何B,C,D品項在2021/10有銷售紀錄，但於2021/11剛好退貨前一個月的銷售數量?

```python
df[df.loc[:,"店家代號"].isin(["868","950","966","967"])]
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 建立畫布和座標
fig, ax = plt.subplots(figsize=(8, 6))

# 設置點的大小
# 假設離群值為大於平均值+3倍標準差,小於平均值-3被標準差
# 1582*3+560=5302
colors = ['grey' if -4186 < nums < 5302 else 'yellow' for nums in dfbyStore.values]
sizes = [1 if  -4186 < nums < 5302 else 300 for nums in dfbyStore.values]
ax.scatter(dfbyStore.index, dfbyStore.values, s=sizes, c=colors, alpha=0.2)

# 新增店家代號標籤
for i, txt in enumerate(dfbyStore.index):
    ax.annotate(txt, (dfbyStore.index[i], dfbyStore.values[i]), ha='center')
    
# 設置label在左上角    
plt.text(5.9,27000, 'std=1582\nmean=560', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))    

# 清空X軸上的刻度,不然太雜
plt.xticks([])
ax.set_ylim([0, 30000])

# 新增x,y的名稱
ax.set_xlabel('store_code')
ax.set_ylabel('sales')
ax.set_title('TotalByStore')

# 匯出
plt.savefig("TotalByStore")

# 顯示圖表
plt.show()
```
