# 2021年總銷售，2022年1-9月總銷售，各品項成長率

```python
# 先把品項column 抓出
Iteam = df.loc[:,"品項"]

# 才能把其他的object型態改成int
dfTotalByItem = df.loc[:,"2022/01":].apply(pd.to_numeric)

# 再concat回來
dfTotalByItem = pd.concat([Iteam,dfTotalByItem], axis="columns")

# 計算2021-2022/09總銷量
dfTotalByItem.groupby("品項").sum().sum(axis="columns")

# 建立DataFrame
outline = pd.DataFrame(dfTotalByItem.groupby("品項").sum().sum(axis="columns"),columns=["2021-2022/09"])
```

```python
# 2021全年各品項銷售總額
dfTotalByItem2021 = dfTotalByItem.groupby("品項").sum().loc[:,"2021/01":"2021/12"].sum(axis="columns")

# 202101~09各品項銷售總額
dfTotalByItem202101_09 = dfTotalByItem.groupby("品項").sum().loc[:,"2021/01":"2021/09"].sum(axis="columns")

# 202201~09各品項銷售總額
dfTotalByItem202201_09 = dfTotalByItem.groupby("品項").sum().loc[:,"2022/01":"2022/09"].sum(axis="columns")

# 202101~08各品項銷售總額
dfTotalByItem202101_08 = dfTotalByItem.groupby("品項").sum().loc[:,"2021/01":"2021/08"].sum(axis="columns")

# 202201~08各品項銷售總額
dfTotalByItem202201_08 = dfTotalByItem.groupby("品項").sum().loc[:,"2022/01":"2022/08"].sum(axis="columns")

# 2022與2021各品項餘01~08月的銷售差額
dfTotalByItem2022_2021 = dfTotalByItem202201_08 - dfTotalByItem202101_08

# 2022與2021各品項餘01~08月的銷售成長率
growth_rate = (dfTotalByItem2022_2021/dfTotalByItem202101_08).apply(lambda x: '{:.2%}'.format(x))


outline["2021"] = dfTotalByItem2021
outline["2021/01-09"] = dfTotalByItem202101_09
outline["2022/01-09"] = dfTotalByItem202201_09
outline["2021/01-08"] = dfTotalByItem202101_08
outline["2022/01-08"] = dfTotalByItem202201_08
outline["01~08_Gap"] = dfTotalByItem2022_2021
outline["GrowthRate"] = growth_rate
outline
```

```python
# 各品項202101-08各月份的銷售額
dfTotalByM_202101_08 =  outlineByMonth.loc[:,"2021/01":"2021/08"]

# 各品項202201-08各月份的銷售額
dfTotalByM_202201_08 =  outlineByMonth.loc[:,"2022/01":"2022/08"]

# 相減
dfGap2022_2021 = dfTotalByM_202201_08 - dfTotalByM_202101_08.values

# 改columns 分別對應月份
dfGap2022_2021.columns = ["1月","2月","3月","4月","5月","6月","7月","8月"]
```

# ItemA

```python
import matplotlib.pyplot as plt
import numpy as np

months = dfTotalByM_202101_08.columns
sales_a = dfTotalByM_202101_08.loc["A"]
sales_b = dfTotalByM_202201_08.loc["A"]

# sales_c = Gap
sales_c = sales_b.values - sales_a.values

# 設置長條圖的位置和寬度
bar_width = 0.6
bar_positions_a = np.arange(len(months))
bar_positions_b = bar_positions_a

# 繪製長條圖
plt.bar(bar_positions_a, sales_a, width=bar_width, alpha=0.5 ,label='2021')
plt.bar(bar_positions_b, sales_b, width=bar_width ,alpha=0.5, label='2022')

# # 在每個長條圖上標註文字

for i in range(len(months)):
    if sales_c[i] >0:
        plt.text(bar_positions_b[i], sales_b[i]+300,"{:.2%}".format(sales_c[i]/sales_a[i]),ha='center')
    else:
        plt.text(bar_positions_b[i], sales_a[i]+300,"{:.2%}".format(sales_c[i]/sales_a[i]),ha='center')

# 設置label在右上角    
plt.text(5.9,28000, 'growth rate', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
    
# 設置標題和軸標籤
plt.title('ItemA ')
plt.xlabel('Month')
plt.ylabel('Sales')

# 設置x軸刻度標籤
plt.xticks(bar_positions_a + bar_width/2, months)

# 更改x軸標籤
new_labels = ['01', '02', '03', '04','05','06','07','08']
plt.xticks(np.arange(len(months)), new_labels)

# 設置圖例
plt.legend()

# 匯出
plt.savefig("ItemA_2022-2021.png")

# 顯示圖形
plt.show()
```
