## 品項A

```python
df[df.loc[:,"品項"] == "A"].loc[:,"2022/01":].apply(pd.to_numeric).describe()
```

```python
totalByY_2021_A = df[df.loc[:,"品項"] == "A"].loc[:,"2021/01":"2021/12"].apply(pd.to_numeric).sum().sum()
print(f'2021年品項A總銷售額: {totalByY_2021_A}')
totalByM_1_8_2021_A = df[df.loc[:,"品項"] == "A"].loc[:,"2021/01":"2021/08"].apply(pd.to_numeric).sum().sum()
print(f'2021年品項A,1~8月的總銷售額: {totalByM_1_8_2021_A}')
totalByM_1_8_2022_A = df[df.loc[:,"品項"] == "A"].loc[:,"2022/01":"2022/08"].apply(pd.to_numeric).sum().sum()
print(f'2022年品項A,1~8月的總銷售額: {totalByM_1_8_2022_A}')

# 品項A 2022與2021年的銷售額差距 , 1-4月和8月正成長 5-7月
A_2022 = df[df.loc[:,"品項"] == "A"].loc[:,"2022/01":"2022/08"].apply(pd.to_numeric).sum().values
A_2021 = df[df.loc[:,"品項"] == "A"].loc[:,"2021/01":"2021/08"].apply(pd.to_numeric).sum().values
pd.DataFrame(A_2022 - A_2021 , index = ["1月","2月","3月","4月","5月","6月","7月","8月"]).T
```

```python
df[df.loc[:,"品項"] == "A"].loc[:,"2021/01":"2021/12"].apply(pd.to_numeric).sum()
```

```python
import matplotlib.pyplot as plt
import numpy as np

months = df.loc[:,"2021/01":"2021/08"].columns
sales_a = df[df.loc[:,"品項"] == "A"].loc[:,"2021/01":"2021/08"].apply(pd.to_numeric).sum()
sales_b = df[df.loc[:,"品項"] == "A"].loc[:,"2022/01":"2022/08"].apply(pd.to_numeric).sum()
sales_c = sales_b.values - sales_a.values

# 設置長條圖的位置和寬度
bar_width = 0.6
bar_positions_a = np.arange(len(months))
bar_positions_b = bar_positions_a

# 繪製長條圖
plt.bar(bar_positions_a, sales_a, width=bar_width, alpha=0.5 ,label='2021')
plt.bar(bar_positions_b, sales_b, width=bar_width ,alpha=0.5, label='2022')

# # 在每個長條圖上標註文字

for i in range(len(months)):
    if sales_c[i] >0:
        plt.text(bar_positions_b[i], sales_b[i]+300,"{:.2%}".format(sales_c[i]/sales_a[i]),ha='center')
    else:
        plt.text(bar_positions_b[i], sales_a[i]+300,"{:.2%}".format(sales_c[i]/sales_a[i]),ha='center')

# 設置label在右上角    
plt.text(5.9,28000, 'growth rate', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
    
# 設置標題和軸標籤
plt.title('ItemA ')
plt.xlabel('Month')
plt.ylabel('Sales')

# 設置x軸刻度標籤
plt.xticks(bar_positions_a + bar_width/2, months)

# 更改x軸標籤
new_labels = ['01', '02', '03', '04','05','06','07','08']
plt.xticks(np.arange(len(months)), new_labels)

# 設置圖例
plt.legend()

# 匯出
plt.savefig("ItemA_2022-2021.png")

# 顯示圖形
plt.show()


```
