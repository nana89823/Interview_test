```python
import pandas as pd
pd.set_option("display.max_columns",30)
pd.set_option("display.max_rows",6000)
```

```python
df = pd.read_csv("大數據分析應用管理師_測試題目_林哲緯_0312 - data set.csv",header=None)
df = df.loc[2:,0:]
df.head()
```

```python
df.dropna(subset=0)
```

```python
df.loc[3:913,0]="A"
df.loc[914:1814,0]="B"
df.loc[1815:2701,0]="C"
df.loc[2702:3544,0]="D"
df.loc[3545:3799,0]="E"
df.loc[3800:4360,0]="F"
df.loc[4361:4965,0]="G"
df.loc[4966:,0]="H"
```

```python
df = df.fillna(0)
df.head()
```

```python
df1 = df.loc[2,0:]
# 把cloumns 名稱改為時間
df.columns= df1.tolist()
```

```python
df = df.loc[3:,:]
```

```python
df.head()
```

*  各店的A品項三種時間的總銷售額
  + 2021
  + 2021/01-08
  + 2022/01-08
* 各品項2022與2021年的銷售額差距

```python
totalByStore_2021_A = df[df.loc[:,"品項"] == "A"].loc[:,"2021/01":"2021/12"].apply(pd.to_numeric).sum(axis=1)
totalByStore_1_8_2021_A = df[df.loc[:,"品項"] == "A"].loc[:,"2021/01":"2021/08"].apply(pd.to_numeric).sum(axis=1)
totalByStore_1_8_2022_A = df[df.loc[:,"品項"] == "A"].loc[:,"2022/01":"2022/08"].apply(pd.to_numeric).sum(axis=1)
Gap = totalByStore_1_8_2022_A - totalByStore_1_8_2021_A

# # 把index改成店家代號,才能判斷是哪間賣的
totalByStore_2021_A.index = list(df.loc[3:913,"店家代號"])
totalByStore_1_8_2021_A.index = list(df.loc[3:913,"店家代號"])
totalByStore_1_8_2022_A.index = list(df.loc[3:913,"店家代號"])
Gap.index = list(df.loc[3:913,"店家代號"])
dfA = pd.DataFrame(totalByStore_2021_A ,columns=["2021"])
dfA["2021/01-08"] = totalByStore_1_8_2021_A
dfA["2022/01-08"] = totalByStore_1_8_2022_A
dfA["Gap"] = Gap
```

```python
dfA
```

```python
dfA.describe()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 建立畫布和座標
fig, ax = plt.subplots(figsize=(8, 6))

# 設置點的大小
# 假設離群值為大於平均值+3倍標準差,小於平均值-3被標準差
# 142*3+3=429
colors = ['grey' if -423 < gap < 429 else 'red' for gap in dfA.loc[:,"Gap"]]
sizes = [1 if  -423 < gap < 429 else 300 for gap in dfA.loc[:,"Gap"]]
ax.scatter(dfA.index, dfA.loc[:,"Gap"], s=sizes, c=colors, alpha=0.2)

# 新增店家代號標籤
for i, txt in enumerate(dfA.index):
    ax.annotate(txt, (dfA.index[i], dfA.loc[:,"Gap"][i]), ha='center')
    
# 設置label在左上角    
plt.text(5.9,1600, 'std=142\nmean=3', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))    

# 清空X軸上的刻度,不然太雜
plt.xticks([])

# 新增x,y的名稱
ax.set_xlabel('store_code')
ax.set_ylabel('gap')
ax.set_title('itemA')

# 匯出
plt.savefig("ItemA_sotre")

# 顯示圖表
plt.show()
```
