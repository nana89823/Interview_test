```python
totalByY_2021 = dfA.loc[:,"2021年1月":].sum().sum()
totalByM_2021 = dfA.loc[:,"2021年1月":].sum()
totalByM_1_9_2021 = dfA.loc[:,"2021年1月":"2021年9月"].sum().sum()
print(f'2021年各月份的銷售額:\n{totalByM_2021}')
print("------")
print(f'2021年總共的銷售額:\n{totalByY_2021}')
print("------")
print(f'2021年1~9月的總銷售額:\n{totalByM_1_9_2021}')
```

```python
totalByY_1_9_2022 = dfA.loc[:,"2022年1月":"2022年9月"].sum()
totalByM_1_9_2022 = dfA.loc[:,"2022年1月":"2022年9月"].sum().sum()
print(f'2022年1~9月的銷售額:\n{totalByY_1_9_2022}')
print("------")
print(f'2021年1~9月的總銷售額:\n{totalByM_1_9_2022}')
```
