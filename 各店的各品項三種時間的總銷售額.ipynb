```python
df = df.loc[3:,:]
```

*  各店的A品項三種時間的總銷售額
  + 2021
  + 2021/01-08
  + 2022/01-08
* 各品項2022與2021年的銷售額差距

```python
totalByStore_2021_A = df[df.loc[:,"品項"] == "A"].loc[:,"2021/01":"2021/12"].apply(pd.to_numeric).sum(axis=1)
totalByStore_1_8_2021_A = df[df.loc[:,"品項"] == "A"].loc[:,"2021/01":"2021/08"].apply(pd.to_numeric).sum(axis=1)
totalByStore_1_8_2022_A = df[df.loc[:,"品項"] == "A"].loc[:,"2022/01":"2022/08"].apply(pd.to_numeric).sum(axis=1)

# 把index改成店家代號,才能判斷是哪間賣的
totalByStore_2021_A.index = list(df.loc[3:913,"店家代號"])
totalByStore_1_8_2021_A.index = list(df.loc[3:913,"店家代號"])
totalByStore_1_8_2022_A.index = list(df.loc[3:913,"店家代號"])
dfA = pd.DataFrame(totalByStore_2021_A ,columns=["2021"])
dfA["2021/01-08"] = totalByStore_1_8_2021_A
dfA["2022/01-08"] = totalByStore_1_8_2022_A
```
